What is the best Retailer in Sales??

sol : West Gear is the best Retailer in sales then Foot lockerGear  while Foot locker is the most distibuted then west Gear


```python
#@title Display sales and profit by Retailer

retailer_analysis = data.groupby('Retailer')['Total Sales', 'Operating Profit'].sum().reset_index()

# Set a larger figure size
plt.figure(figsize=(16, 8))

# Create a bar chart for retailer distribution
plt.subplot(2, 1, 1)
sns.countplot(x='Retailer', data=data, order=data['Retailer'].value_counts().index)
plt.title('Distribution of Retailers')
plt.xlabel('Retailer')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Create a bar chart for total sales and operating profit for each retailer
plt.subplot(2, 1, 2)
sns.barplot(x='Retailer', y='Total Sales', data=retailer_analysis, color='skyblue', label='Total Sales')
sns.barplot(x='Retailer', y='Operating Profit', data=retailer_analysis, color='orange', label='Operating Profit')
plt.title('Total Sales and Operating Profit by Retailer')
plt.xlabel('Retailer')
plt.ylabel('Amount')

# Set larger y-axis ticks and labels
plt.ticklabel_format(style='plain', axis='y')  # Disable scientific notation
plt.yticks(fontsize=12)  # Adjust y-axis tick font size

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.legend()

# Adjust layout for better visualization
plt.tight_layout()

# Show the plot
plt.show()
```

What is the best Region in Sales and Profit??

sol : South East is the best Region in sales and profit

```python
#@title Display sales and profit by Regions
# Region and State Analysis
region_state_analysis = data.groupby(['Region', 'State'])[['Total Sales', 'Operating Profit']].sum().reset_index()

# Set a larger figure size
plt.figure(figsize=(16, 8))

# Create a bar chart for total sales and operating profit by region
plt.subplot(2, 1, 1)
sns.barplot(x='Region', y='Total Sales', data=region_state_analysis, palette='Blues', ci=None)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.ticklabel_format(style='plain', axis='y')  # Disable scientific notation on the y-axis

# Create a bar chart for operating profit by region
plt.subplot(2, 1, 2)
sns.barplot(x='Region', y='Operating Profit', data=region_state_analysis, palette='Oranges', ci=None)
plt.title('Operating Profit by Region')
plt.xlabel('Region')
plt.ylabel('Operating Profit')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.ticklabel_format(style='plain', axis='y')  # Disable scientific notation on the y-axis

# Adjust layout for better visualization
plt.tight_layout()

# Show the plot
plt.show()
```

What is the top City in sales and profit??

sol : New York is the best City in sales and profit while it is not in the top 2 regions in sales .


```python
#@title Top 10 Cities in sales and profit
city_analysis = data.groupby(['Region', 'City'])[['Total Sales', 'Operating Profit']].sum().reset_index()

# Sort the DataFrame by total sales or operating profit (choose one)
sorted_city_analysis = city_analysis.sort_values(by='Total Sales', ascending=False)
# OR
# sorted_city_analysis = city_analysis.sort_values(by='Operating Profit', ascending=False)

# Select the top 10 cities
top_10_cities = sorted_city_analysis.head(10)

# Set a larger figure size
plt.figure(figsize=(16, 8))

# Define a custom color palette with four different shades of blue
custom_palette = sns.color_palette("Blues", n_colors=4)

# Create a subplot for total sales
plt.subplot(2, 1, 1)
sns.barplot(x='City', y='Total Sales', hue='Region', data=top_10_cities, palette=custom_palette, ci=None)
plt.title('Top 10 Cities - Total Sales by Region')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.ticklabel_format(style='plain', axis='y')  # Disable scientific notation on the y-axis
plt.legend(title='Region', loc='upper right')

# Create a subplot for operating profit
plt.subplot(2, 1, 2)
sns.barplot(x='City', y='Operating Profit', hue='Region', data=top_10_cities, palette=custom_palette, ci=None)
plt.title('Top 10 Cities - Operating Profit by Region')
plt.xlabel('City')
plt.ylabel('Operating Profit')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.ticklabel_format(style='plain', axis='y')  # Disable scientific notation on the y-axis

# Adjust layout for better visualization
plt.tight_layout()

# Show the plot
plt.show()
```

What is the best quarter in the year in sales?

sol: Q3 is the best Quarter in sales

Q4 is the worst   

```python
#@title Top sales by 2020
data['Invoice Date'] = pd.to_datetime(data['Invoice Date'])

# Filter the data for the year 2020
data_2020 = data[data['Invoice Date'].dt.year == 2020]

# Extract the quarter from the 'Invoice Date' column for the year 2020
data_2020['Quarter'] = data_2020['Invoice Date'].dt.to_period('Q')

# Group by quarter and calculate the sum of sales for each quarter
quarterly_sales_2020 = data_2020.groupby('Quarter')['Total Sales'].sum().reset_index()

# Set a larger figure size
plt.figure(figsize=(12, 6))

# Create a bar plot for sales by quarters in 2020
sns.barplot(x='Quarter', y='Total Sales', data=quarterly_sales_2020, ci=None)
plt.title('Sales by Quarters in 2020')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')

# Show the plot
plt.show()
```

What is the most selling Product??

Mens Street Footwear is the most selling product.

```python
#@title Sales by product
product_sales = data.groupby('Product')['Total Sales'].sum().reset_index()

# Sort the DataFrame by total sales in descending order to get top-selling products
top_selling_products = product_sales.sort_values(by='Total Sales', ascending=False)

# Display the top-selling products
print("Top-Selling Products:")
print(top_selling_products.head())

# Plot the top-selling products
plt.figure(figsize=(12, 6))
sns.barplot(x='Total Sales', y='Product', data=top_selling_products, ci=None)
plt.title('Top-Selling Products')
plt.xlabel('Total Sales')
plt.ylabel('Product')
plt.show()
```

```python
sales_method_distribution = data['Sales Method'].value_counts()

# Display the distribution of sales methods
print("Sales Method Distribution:")
print(sales_method_distribution)

# Plot the distribution of sales methods
plt.figure(figsize=(12, 6))
sns.countplot(x='Sales Method', data=data)
plt.title('Sales Method Distribution')
plt.xlabel('Sales Method')
plt.ylabel('Count')
plt.show()

# Analyze the impact of different sales methods on total sales and profit
sales_method_analysis = data.groupby('Sales Method')[['Total Sales', 'Operating Profit']].sum()

# Plot the impact of different sales methods
sales_method_analysis.plot(kind='bar', figsize=(12, 6))
plt.title('Impact of Sales Methods on Total Sales and Profit')
plt.xlabel('Sales Method')
plt.ylabel('Amount')
plt.show()
```
